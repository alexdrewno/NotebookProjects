# World Cities CSV Parser

### I am using data available from https://simplemaps.com/data/world-cities to process the data into a dictionary of cities with keys being the country name

### This output JSON is what is going to be stored in my application Lost On Earth (which is a geo guessing game)

```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv("./worldcities.csv")
```

```python
df
```

```python
#dictionary of cities with countries as keys
countries_dict = {}

for _, row in df.iterrows():
    cities_dict = {}
    cities_dict['city'] = row['city']
    cities_dict['long'] = row['lng']
    cities_dict['lat'] = row['lat']
    
    if row['country'] not in countries_dict:
         countries_dict[row['country']] = []
    
    countries_dict[row['country']].append(cities_dict)
```

```python
with open('parsed_world_cities.json', 'w') as fp:
    json.dump(countries_dict, fp)
```
